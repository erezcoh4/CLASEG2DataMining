```python
def print_output(particle=dict(), p=(1,1,1), v=(0,0,0), t_off=0 , flag=0):
    '''
    p is 3-momentum
    '''
    p_mag = np.sqrt( np.square(p[0]) + np.square(p[1]) + np.square(p[2]) ) 
    p_components = np.array(p)*1./p_mag
    print '2'
    print particle['pid'],' ',p_components[0],' ',p_components[1],' ',p_components[2],' ',p_mag
    print particle['mass'],' ',particle['charge'],' '
    print v[0],' ',v[1],' ',v[2],' ',t_off,' ',flag,' '
    
    
def output2file(particle=dict(), p=(1,1,1), outputfile=None, v=(0,0,0), t_off=0 , flag=0):
    '''
    p is 3-momentum in cartesizan coordinates
    '''
    p_mag = np.sqrt( np.square(p[0]) + np.square(p[1]) + np.square(p[2]) ) 
    p_components = np.array(p)/p_mag
    outputfile.write('%d'%particle['pid']+' %f'%p_components[0]+' %f'%p_components[1]+' %f'%p_components[2]+' %f'%p_mag+'\n')
    outputfile.write('%f'%particle['mass']+' %d'%particle['charge']+'\n')
    outputfile.write('%f'%v[0]+' %f'%v[1]+' %f'%v[2]+' %d'%t_off+' %d'%flag+'\n')
```

```python
def print_event( Pe, Pprotons ):    
    print_output( electron , Pe )
    for Pp in Pprotons:
        print_output( proton , Pp )  
        
def output_event( Pe=None, outputfile=None, Pprotons=None ):    
    # momenta are in spherical coordinates
    if Pprotons is not None:
        Nparticles = 1 + len(Pprotons)
    else:
        Nparticles = 1 
    outputfile.write('%d\n'%Nparticles)
    output2file( electron , Pe , outputfile)
    if Pprotons is not None:
        for Pp in Pprotons:
            output2file( particle=proton , p=Pp , outputfile=outputfile)  
```

```python
hyperparameters = dict()
hyperparameters['run'] = 1006
```

```python
hyperparameters['mode'] = 'smeared electron in center of sector' # 'fixed-electron'
hyperparameters['min p'] = 0.8
hyperparameters['max p'] = 2.4
hyperparameters['min theta'] = 20 # deg.
hyperparameters['max theta'] = 90 # deg.
hyperparameters['min phi'] = 0 # deg.
hyperparameters['max phi'] = 360 # deg.
for angle in ['phi','theta']:
    hyperparameters['min '+angle+' rad'] = (np.pi/180.)*hyperparameters['min '+angle]
    hyperparameters['max '+angle+' rad'] = (np.pi/180.)*hyperparameters['max '+angle]
hyperparameters['max cos theta'] = np.cos(hyperparameters['min theta rad']) 
hyperparameters['min cos theta'] = np.cos(hyperparameters['max theta rad']) 
hyperparameters['N events'] = 100000
print 'hyperparameters:\n',hyperparameters
```

```python
electron = dict({'mass':0.000511,'charge':-1,'pid':11})
proton = dict({'mass':0.938272,'charge':1,'pid':2212})
```

```python
generated_df = pd.DataFrame()
outputfile = open('/Users/erezcohen/Desktop/DataMining/GSIM/eg_txtfiles/run%d.txt'%hyperparameters['run'],'w')
for i_event in range(hyperparameters['N events']):
    if i_event%(hyperparameters['N events']/10)==0: print '%.0f'%(100*float(i_event)/hyperparameters['N events'])+'%'
        
    if hyperparameters['mode'] == 'fixed-electron':
        Pe_fixed = (-0.137*4.306 , -0.339*4.306 , 0.956*4.306 ) 
        Pe = Pe_fixed
        Pe_x = Pe[0]
        Pe_y = Pe[1]
        Pe_z = Pe[2]
        Pe_P = np.sqrt(np.square(Pe_x)+np.square(Pe_y)+np.square(Pe_z))
        Pe_theta = np.arccos(Pe_z/Pe_P)
        Pe_phi = np.arctan2(Pe_y,Pe_x)

    if hyperparameters['mode'] == 'smeared electron in center of sector':
        Pe_P = np.random.uniform(2,2.5)
        cos_Pe_theta = np.random.uniform(0.7986,0.8387)
        Pe_theta = np.arccos(cos_Pe_theta)
        Pe_phi = np.random.uniform(1.012,1.082)
        Pe_x = Pe_P*np.sin(Pe_theta)*np.cos(Pe_phi)
        Pe_y = Pe_P*np.sin(Pe_theta)*np.sin(Pe_phi)
        Pe_z = Pe_P*cos_Pe_theta
        Pe = ( Pe_x, Pe_y , Pe_z  )

        
    # proton from a uniform random distribution
    Pp_P = np.random.uniform(hyperparameters['min p'],hyperparameters['max p'])
    Pp_phi = np.random.uniform(hyperparameters['min phi rad'],hyperparameters['max phi rad'])
    cos_Pp_theta = np.random.uniform(hyperparameters['min cos theta'],hyperparameters['max cos theta'])
    Pp_theta = np.arccos(cos_Pp_theta)
    Pp_x = Pp_P*np.sin(Pp_theta)*np.cos(Pp_phi)
    Pp_y = Pp_P*np.sin(Pp_theta)*np.sin(Pp_phi)
    Pp_z = Pp_P*cos_Pp_theta
    Pp = ( Pp_x, Pp_y , Pp_z  )

    # output to text file
    output_event( Pe=Pe, Pprotons=[Pp] , outputfile=outputfile )
    
    generated_df=generated_df.append({'Pp_P':Pp_P,'Pp_theta':Pp_theta,'Pp_phi':Pp_phi
                                      ,'Pe_P':Pe_P,'Pe_theta':Pe_theta,'Pe_phi':Pe_phi
                                      ,'Pe_x':Pe_x,'Pe_y':Pe_y,'Pe_z':Pe_z
                                     },ignore_index=True)
    
print
print 'done. wrote',len(generated_df),'events into'
print '/Users/erezcohen/Desktop/DataMining/GSIM/eg_txtfiles/run%d.txt'%hyperparameters['run'] 
```

```python
electron = dict({'mass':0.000511,'charge':-1,'pid':11})
proton = dict({'mass':0.938272,'charge':1,'pid':2212})
```

```python
e_theta_min_data = np.min(dataset['Pe_theta'])
print 'e_theta_min_data:',e_theta_min_data
e_theta_max_data = np.max(dataset['Pe_theta'])
print 'e_theta_max_data:',e_theta_max_data
```

```python
generated_df = pd.DataFrame()
N_events = 300000#62600
outputfile = open('/Users/erezcohen/Desktop/DataMining/GSIM/eg_txtfiles/run%d.txt'%run,'w')
for i_event in range(N_events):
    if i_event%(N_events/10)==0: print '%.0f'%(100*float(i_event)/N_events)+'%'
    # recoil proton from a random data event
    i_data_event = np.random.randint(N_data_events)
    data_event = dataset.iloc[[i_data_event]]
    Prec_x = float(data_event['Prec_x'])
    Prec_y = float(data_event['Prec_y'])
    Prec_z = float(data_event['Prec_z'])    
    Prec = ( Prec_x, Prec_y, Prec_z  )
    Prec_P = np.sqrt(np.square(Prec_x)+np.square(Prec_y)+np.square(Prec_z))
    Prec_theta = np.arccos(Prec_z/Prec_P)
    Prec_phi = np.arctan2(Prec_y,Prec_x)


    if mode == 'variable-electron':
        # variable electron
        # an electron, generated from the statistical behavior of the electon-Precoil correlations
        # namely, the electron - Prec relation comes from another random data event
        Pe_theta = 0
        while Pe_theta<e_theta_min_data or Pe_theta>e_theta_max_data:
            i_data_event = np.random.randint(N_data_events)
            data_event = dataset.iloc[[i_data_event]]
            Pe_theta = Prec_theta + float(data_event['DPePrec_theta'])

        Pe_phi = Prec_phi + float(data_event['DPePrec_phi'])
        Pe_P = Prec_P + float(data_event['DPePrec_P'])
        if Pe_P<0: print "float(data_event['DPePrec_P']):",float(data_event['DPePrec_P'])
        Pe_x = Pe_P*np.sin(Pe_theta)*np.cos(Pe_phi)
        Pe_y = Pe_P*np.sin(Pe_theta)*np.sin(Pe_phi)
        Pe_z = Pe_P*np.cos(Pe_theta)
        Pe = ( Pe_x , Pe_y , Pe_z )
        
    elif mode == 'fixed-electron':
        # fixed electron
        Pe = Pe_fixed
        Pe_x = Pe[0]
        Pe_y = Pe[1]
        Pe_z = Pe[2]
        Pe_P = np.sqrt(np.square(Pe_x)+np.square(Pe_y)+np.square(Pe_z))
        Pe_theta = np.arccos(Pe_z/Pe_P)
        Pe_phi = np.arctan2(Pe_y,Pe_x)


    
    # monitor the electron-Precoil correlations
    DPePrec_phi = Prec_phi - Pe_phi
    DPePrec_theta = Prec_theta - Pe_theta

    # output to text file
    output_event( Pe=Pe, Pprotons=[Prec] , outputfile=outputfile )
    
    generated_df=generated_df.append({'Prec_P':Prec_P,'Prec_theta':Prec_theta,'Prec_phi':Prec_phi
                                      ,'Pe_P':Pe_P,'Pe_theta':Pe_theta,'Pe_phi':Pe_phi
                                      ,'Pe_x':Pe_x,'Pe_y':Pe_y,'Pe_z':Pe_z
                                      ,'DPePrec_phi':DPePrec_phi,'DPePrec_theta':DPePrec_theta
                                     },ignore_index=True)
    
print
print 'done. wrote',len(generated_df),'events into'
print '/Users/erezcohen/Desktop/DataMining/GSIM/eg_txtfiles/run%d.txt'%run 
```

```python
electron = dict({'mass':0.000511,'charge':-1,'pid':11})
proton = dict({'mass':0.938272,'charge':1,'pid':2212})
Pp_min,Pp_max = 0.2,1.0
```

```python
# a single electron that passes RECSIS cuts...
Pe = (-0.137*4.306 , -0.339*4.306 , 0.956*4.306 ) 
```

```python
generated_df = pd.DataFrame()
N_events = 1000 #000
outputfile = open('/Users/erezcohen/Desktop/DataMining/GSIM/eg_txtfiles/run%d.txt'%run,'w')
for i_event in range(N_events):
    if i_event%(N_events/10)==0: print '%.0f'%(100*float(i_event)/N_events)+'%'
    # unifrom proton momentum magnitude in the desired range
    Pp_mag = np.random.uniform( Pp_min , Pp_max )
    # unifrom proton direction
    Pp_cos_theta = np.random.uniform( -1 , 1 )
    Pp_theta = np.arccos(Pp_cos_theta)
    Pp_phi = np.random.uniform( 0 , 2*np.pi )
    
    Pp = ( Pp_mag*np.sin(Pp_theta)*np.cos(Pp_phi), Pp_mag*np.sin(Pp_theta)*np.sin(Pp_phi), Pp_mag*np.cos(Pp_theta) )
    #print_event( Pe, [Pp] )
    output_event( Pe, [Pp] , outputfile )
    
    generated_df=generated_df.append({'Pp_mag':Pp_mag,'Pp_theta':Pp_theta,'Pp_phi':Pp_phi
                                     },ignore_index=True)
    
print 'done. wrote\n'+'/Users/erezcohen/Desktop/DataMining/GSIM/eg_txtfiles/run%d.txt'%run 
```
