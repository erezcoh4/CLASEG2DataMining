```python
acceptance = pd.read_csv('output.csv')
```

```python
fig=plt.figure(figsize=(24,8))
ax = fig.add_subplot(1,3,1)
plt.hist(acceptance.N_generated_in_event,bins=24);
set_axes(ax,r'$N_{generated}$')

ax = fig.add_subplot(1,3,2)
plt.hist(acceptance.e_accepted_in_event,bins=15);
set_axes(ax,r'$N_{e \; accepted}$')

ax = fig.add_subplot(1,3,3)
plt.hist(acceptance.ep_accepted_in_event,bins=12);
set_axes(ax,r'$N_{e+p \; accepted}$')
```

```python
e_acceptance = np.array(acceptance['e_accepted_in_event']/acceptance['N_generated_in_event'])
ep_acceptance = np.array(acceptance['ep_accepted_in_event']/acceptance['N_generated_in_event'])
```

```python
ep_acceptance_e_acceptance = [ep_acceptance[i]/(e_acceptance[i] if e_acceptance[i]>0 else 1)  for i in range(len(e_acceptance))]
```

```python
fig=plt.figure(figsize=(24,8))
ax = fig.add_subplot(1,3,1)
plt.hist(e_acceptance);
set_axes(ax,r"(e,e') acceptance")

ax = fig.add_subplot(1,3,2)
plt.hist(ep_acceptance);
set_axes(ax,r"(e,e'p) acceptance")

ax = fig.add_subplot(1,3,3)
plt.hist(ep_acceptance_e_acceptance);
set_axes(ax,r"$\frac{(e,e'p) \; acceptance}{(e,e') \; acceptance}$")
```
