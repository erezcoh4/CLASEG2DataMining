```python
import numpy as np, pandas as pd
```

```python
x = np.linspace(0,2,100)
dx = [0.1 for i in range(len(x))]
y = np.random.uniform(0,1,100)
dy = [0.01 for i in range(len(y))]
```

```python
df = pd.DataFrame(columns=['x','y','dx','dy'])
```

```python
for x_,y_,dx_,dy_ in zip(x,y,dx,dy):
    df_tmp = pd.DataFrame({'x':x_,'y':y_,'dx':dx_,'dy':dy_},index=[0])
    df = df.append(df_tmp)
```

```python
df
```

```python
df = pd.DataFrame({'x':[x_ for x_ in x]
                   ,'y':[y_ for y_ in y]}
                  ,index=[0])
```
