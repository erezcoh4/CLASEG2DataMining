```python
def plot_weighted_variable(data=sims500,var='SigmaX',x_label='$\sigma_t$ [GeV/c]',bins=15):
    weights = data['PvalTotal_allPvals_12C']
#     weights = data['PvalTotal_allPvals_12C']

    fig,ax=plt.subplots(figsize=(8,6))
    h,bins,_= plt.hist(data['gen'+var], bins=bins, histtype='step', linewidth=3, normed=1);
    h,bins,_= plt.hist(data['gen'+var], weights=weights , bins=bins, histtype='step', linewidth=3, normed=1);
    set_axes(ax,x_label=x_label, y_label='')
    mean = np.average(data['gen'+var], weights=weights)
    std = np.sqrt( np.average( np.square(data['gen'+var] - mean), weights=weights ) )
    ax.set_title('weighted average: $%.3f \pm %.3f$'%(mean,std) , fontsize=20,y=1.02)
```

```python
plot_weighted_variable(data=sims500,var='SigmaX',x_label='$\sigma_t$ [GeV/c]',bins=15)
plot_weighted_variable(data=sims500,var='SigmaZa1',x_label='$a_1$ ',bins=15)
plot_weighted_variable(data=sims500,var='SigmaZa2',x_label='$a_2$ [GeV/c]',bins=15)
plot_weighted_variable(data=sims500,var='MeanZa1',x_label='$b_1$',bins=15)
plot_weighted_variable(data=sims500,var='MeanZa2',x_label='$b_2$ [GeV/c]',bins=15)
```
