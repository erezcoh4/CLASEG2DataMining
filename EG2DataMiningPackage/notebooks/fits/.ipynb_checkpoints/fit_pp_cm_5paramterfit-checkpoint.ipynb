# Example (1): 2-dimnesional Gaussian fit of synthetic data 

```python
def Gaussian2D(( x , y), amplitude, mean_x , mean_y , sigma_x, sigma_y  ):
    mean_x , mean_y = float(mean_x) , float(mean_y) 
    amplitude = float(amplitude)
    ax = ((x - mean_x)**2)/(2*sigma_x**2)
    ay = ((y - mean_y)**2)/(2*sigma_y**2)
    g = ( amplitude / (2*sigma_x*sigma_y) ) * np.exp( - (ax + ay) )
    return g.ravel()

# Create x and y indices
x = np.linspace(0, 100, 101)
y = np.linspace(0, 100, 101)
x, y = np.meshgrid(x, y)

# create data
data = Gaussian2D((x, y), 1, 50, 50, 12, 20)
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(1,2,1)
ax.imshow(data.reshape(101, 101), cmap=plt.cm.jet, origin='bottom', extent=(x.min(), x.max(), y.min(), y.max()));

# add some noise to the data 
data_noisy = data + 0.0002*np.random.normal(size=data.shape)

# fit the generated data
initial_guess = (1, 50 , 50, 12, 20)
popt, pcov = opt.curve_fit(Gaussian2D, (x, y), data_noisy, p0=initial_guess)

data_fitted = Gaussian2D((x, y), *popt)
ax = fig.add_subplot(1,2,2)
im = ax.imshow(data_noisy.reshape(101, 101), cmap=plt.cm.jet, origin='bottom',
    extent=(x.min(), x.max(), y.min(), y.max()))
print 'data_fitted.reshape(101, 101):\n',data_fitted.reshape(101, 101)
ax.contour(x, y, data_fitted.reshape(101, 101), 8, colors='w')

amplitude, mean_x , mean_y , sigma_x, sigma_y = popt
print 'amplitude:',amplitude, '\nmean_x:',mean_x , '\nmean_y:',mean_y , '\nsigma_x:',sigma_x, '\nsigma_y:',sigma_y 
print pcov
```

# Example (2): 2-dimnesional Gaussian with an offset of the mean and the width in the y direction

```python
def projection_xy(data):
    data_xy = np.zeros((N,N))
    for i in range(N):
        data_xy += data.reshape(N, N, N)[i]
    return data_xy
```

```python
def Gaussian2DwithOffset(( x , y, Pmiss), amplitude , mean_x  , b1 , b2 , sigma_x , a1 , a2 ):
    amplitude = float(amplitude)
    sigma_y = a1*(Pmiss - 0.6) + a2
    mean_y = b1*(Pmiss - 0.6 ) + b2
    ax = (np.square(x - mean_x))/(2*np.square(sigma_x))
    ay = (np.square(y - mean_y))/(2*np.square(sigma_y))
    g = ( amplitude / (2*sigma_x*sigma_y) ) * np.exp( - (ax + ay) )
    return g.ravel()

# Create x and y indices
N = 4
x = np.linspace(-0.5, 0.5, N)
y = np.linspace(-0.5, 0.5, N)
Pmiss = np.linspace( 0.3 , 1 , N)
x, y , Pmiss = np.meshgrid(x, y, Pmiss)

# create data
data = Gaussian2DwithOffset((x, y, Pmiss), amplitude=1, mean_x=0, b1=0.562 , b2=0.159, sigma_x=0.155, a1=0.16 , a2=0.159)
data_xy = projection_xy(data)
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(1,3,1)
ax.imshow(data_xy, cmap=plt.cm.jet, origin='bottom', extent=(x.min(), x.max(), y.min(), y.max()));


# add some noise to the data 
data_noisy = data + np.random.normal(size=data.shape)
data_noisy_xy = projection_xy(data_noisy)

ax = fig.add_subplot(1,3,2)
im = ax.imshow(data_noisy_xy, cmap=plt.cm.jet, origin='bottom',extent=(x.min(), x.max(), y.min(), y.max()))

# fit the generated data
initial_guess = (1, 0, 0.562 , 0.159, 0.155, 0.16 , 0.159)
popt, pcov = opt.curve_fit(Gaussian2DwithOffset, (x, y, Pmiss), data_noisy, p0=initial_guess)
im = ax.imshow(data_noisy_xy, cmap=plt.cm.jet, origin='bottom',extent=(x.min(), x.max(), y.min(), y.max()))

data_fitted = Gaussian2DwithOffset((x, y, Pmiss), *popt)
data_fitted_xy = projection_xy(data_fitted)

ax = fig.add_subplot(1,3,3)
im = ax.imshow(data_fitted_xy, cmap=plt.cm.jet, origin='bottom',extent=(x.min(), x.max(), y.min(), y.max()))

A, mean_x, b1, b2 , sigma_x , a1 , a2 = popt
A_err, mean_x_err, b1_err, b2_err , sigma_x_err , a1_err , a2_err = pcov[0,0], pcov[1,1], pcov[2,2], pcov[3,3], pcov[4,4], pcov[5,5], pcov[6,6]
results = pd.DataFrame({'A':A,'A_err':A_err,
                        'mean_x':mean_x,'mean_x_err':mean_x_err,
                        'b1':b1, 'b2':b2 ,'b1_err':b1_err, 'b2_err':b2_err ,
                        'sigma_x':sigma_x ,'sigma_x_err':sigma_x_err ,
                        'a1':a1 , 'a2':a2,'a1_err':a1_err , 'a2_err':a2_err,
                        'covariance':[pcov]},index=[0])
results
```

```python
def Gaussian3DwithOffset( x , amplitude , sigma_t , a1 , a2 , b1 , b2 ):
    PcmX , PcmY , PcmZ , Pmiss  = x[0] , x[1] , x[2] , x[3]
    amplitude = float(amplitude)
    mean_x = mean_y = 0
    sigma_x = sigma_t
    sigma_y = sigma_t 
    sigma_z = a1*(Pmiss - 0.6) + a2
    mean_z = b1*(Pmiss - 0.6 ) + b2
    ax = (np.square(PcmX - mean_x))/(2*np.square(sigma_x))
    ay = (np.square(PcmY - mean_y))/(2*np.square(sigma_y))
    az = (np.square(PcmZ - mean_z))/(2*np.square(sigma_z))
    g = ( amplitude / (sqrt2pi*sigma_x*sqrt2pi*sigma_y*sqrt2pi*sigma_z) ) * np.exp( - (ax + ay + az) )
    return g
```

```python
import scipy.optimize as optimize
N     = 100
mu    = 10.0
sigma = 5.0
data = np.random.normal(mu, sigma, N)

def nll(p, D):
    curr_mu    = p[0]
    curr_sigma = p[1]
    ll = 0.
    for x in D:
        residual = x - curr_mu       
        ll      += stats.norm.logpdf(residual, 0, curr_sigma)    
        # Alternatively, using stats.norm.logpdf(x, curr_mu, curr_sigma) is shorter & equivalent,
        # but doing it in terms of the residual is a more direct analog to this week's homework.
    return -ll
p0 = np.array([ 0.0, 1.0 ])  # start at a standard normal; anything will do.
result    = optimize.minimize(nll, p0, (data), bounds = [(None,None), (0.0,None)])
if result.success != True:
    sys.exit("Maximum likelihood fit failed")
mu_fit    = result.x[0]
sigma_fit = result.x[1]


print("mean = {0:.2f}".format(np.mean(data)))
print("sd   = {0:.2f}".format(np.std(data)))
print result
print("ML fit, mu    = {0:.2f}".format(mu_fit))
print("ML fit, sigma = {0:.2f}".format(sigma_fit))
```

```python
def projection_xyz(data,N):
    data_xyz = np.zeros((N,N,N))
    for i in range(N):
        data_xyz += data.reshape(N, N, N, N)[i]
    return data_xyz
def projection_x(data,N):
    data_xy = projection_xy(data)
    data_x = np.zeros(N)
    for i in range(N):
        data_x += data_xy.reshape(N, N)[i]
    return data_x

def project_to_xyz(array_4d):
    array_xyz = np.zeros((len(array_4d),len(array_4d),len(array_4d)))
    for i in range(len(array_4d)):
        array_xyz += array_4d[i]
    return  array_xyz

def project_to_xy(array_xyz):
    array_xy = np.zeros((len(array_xyz),len(array_xyz)))
    for i in range(len(array_xyz)):
        array_xy += array_xyz[i]
    return  array_xy

def project_to_x(project_to_xy):
    array_x = np.zeros(len(array_xy))
    for i in range(len(array_xy)):
        array_x += array_xy[i]
    return  array_x
```
