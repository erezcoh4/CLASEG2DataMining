```python
def label_pcm(dir='x',units='GeV/c'):
    return '$p_{c.m.}^{'+dir+'}$ '+ ('['+units+']') if units!='' else ''

def label_pmiss(dir='',units='GeV/c'):
    return '$p_{miss}^{'+dir+'}$ '+ ('['+units+']') if units!='' else ''
```

# pp-SRC c.m. analysis

```python
#with offset 0.6
columns = ['Q2','theta (miss-q)','Pmiss3Mag','pcmX','pcmY','pcmZ']
data = dict()
for target in ['C12','Al27','Fe56','Pb208']:
    data[target] = read_root( path+"/AnaFiles/Ana_ppSRCcut_DATA_%s.root"%target, columns=columns )
```

```python
target = 'C12'
fig = plt.figure(figsize=(16,8))
bins = 25
ax = fig.add_subplot(121)
plot_2d_hist(data[target].pcmX,data[target].pcmY,cmap='hot_r',bins=bins
             ,xlabel=label_pcm('x'),ylabel=label_pcm('y'), ax=ax);

ax = fig.add_subplot(122)
plot_2d_hist(data[target].pcmZ,data[target].Pmiss3Mag,cmap='hot_r',bins=bins
             ,xlabel=label_pcm('\\vec{p}_{miss}'),ylabel=label_pmiss(), ax=ax);
plt.tight_layout()
```
