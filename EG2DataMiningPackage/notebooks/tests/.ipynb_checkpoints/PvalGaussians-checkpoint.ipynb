# Test Pvalue combinations

```python
N = 10000
Pval_1 , Pval_2 , Pval_3, Pval_Fisher_12, Pval_Fisher_123 , Pval_Fisher_12323 = [] , [] , []  , [] , [] , []
for i in range(N):
    if i%(N/10)==0: print (100*float(i)/N),'%'
        
    P1 = np.random.normal(0.147,0.013)
    Pval_1.append(P1)
    
    P2 = np.random.normal(0.158,0.015)
    Pval_2.append(P2)

    P3 = np.random.normal(0.172,0.018)
    Pval_3.append(P3)

    
    PFisher_12 = Fisher_combination_Pvals([P1,P2])
    Pval_Fisher_12.append(PFisher_12)

    PFisher_123 = Fisher_combination_Pvals([P1,P2,P3])
    Pval_Fisher_123.append(PFisher_123)

    PFisher_12323 = Fisher_combination_Pvals([P1,P2,P3,P2,P3])
    Pval_Fisher_12323.append(PFisher_12323)

    
print 'done'
```

```python
fig,ax = plt.subplots(figsize=(10,6))
for x,label in zip([Pval_1,Pval_2,Pval_3,Pval_Fisher_123]
                   ,['$Pval_1$','$Pval_2$','$Pval_3$','Fisher $Pval_{1,2,3}$']):
    plt.hist(x,bins=np.linspace(0,1,50),label=label,histtype='step',linewidth=3,normed=1)
# ax.set_xscale('log')
# ax.set_yscale('log')    
set_axes(ax,'$Pval$')
ax.legend(fontsize=20)
# ax.xaxis.set_ticks([0.00001,0.0001,0.001,0.01,0.1,1]);
```
