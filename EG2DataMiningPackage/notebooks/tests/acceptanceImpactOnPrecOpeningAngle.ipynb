## Sigma(T)=0.150 Sigma(Z)=0.211 Mean(Z)=0.004

```python
fig=plt.figure(figsize=(16,8))
for i,(var,x_label,xmax) in enumerate(zip(['theta_PleadPrecoil','Precoil3Mag']
                                     ,['$\\theta_{p_{lead},p_{recoil}}$ [deg.]','$p_{recoil}$ [GeV/c]'],[180,1.3])):
    ax=fig.add_subplot(1,2,i+1)
    for sim,label,color in zip([gen_sim1,rec_sim1]
                           ,['generated','reconstructed']              
                           ,['black','blue']):
        plt.hist(sim[var],bins=np.linspace(0,xmax,100),color=color,histtype='step',linewidth=3,label=label)
    plt.legend(fontsize=20,loc='upper left' if i==0 else 'upper right')
    set_axes(ax,x_label=x_label,do_add_grid=True)
plt.tight_layout()
```
