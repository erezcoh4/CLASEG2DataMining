```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
from notebook_tools import *
%matplotlib inline
sigma = np.array([0 , 60, 100 , 158 , 181 , 175])
sigma_err = np.array([0 , 10, 20 , 15 , 15 , 15])

sigma2 = np.square(sigma)
sigma2_err = 2*sigma*sigma_err


a2 = np.array([1, 1.97, 3.80 , 4.75 ,  5.58 , 5.29 ])
a2_err = np.array([0 , 0.1 , 0.1 , 0.25 , 0.25 , 0.25])
```

```python
ColleCalc = pd.DataFrame({'a2':[ 3.80 , 4.75, 5.58 , 5.29 ],
                          'A':[ 4 , 12 , 56 , 208],
                        'allpairs':[110,140,146,145],
                        'nl00':[110,156,172,178]},
                         index = ['$^{4}$He','$^{12}$C','$^{56}$Fe','$^{208}$Pb'])
```

```python
fig,ax=plt.subplots(figsize=(10,6))
plt.errorbar(a2,sigma,xerr=a2_err,yerr=sigma_err,fmt='o',label='data')
p = np.poly1d(np.polyfit(x=a2,y=sigma,deg=1))
plt.plot(a2,p(a2),'--')
plt.plot(ColleCalc['a2'],ColleCalc['allpairs'],'s',color='red',label='Colle, all pairs')
plt.plot(ColleCalc['a2'],ColleCalc['nl00'],'v',color='black',label='Colle, n,l=0,0')
plt.legend(loc='best',fontsize=20)
set_axes(ax,'a2','$\\sigma_{c.m.}$ [MeV/c]')
plt.text(1.8,80,'$^3$He',fontsize=20)
plt.text(3.6,60,'$^4$He',fontsize=20)
plt.text(4.6,110,'$^{12}$C',fontsize=20)
plt.text(5.4,130,'$^{56}$Fe',fontsize=20)
plt.text(5.0,185,'$^{208}$Pb',fontsize=20)
plt.text(1.5,100,'preliminary',fontsize=100,alpha=0.05)
```
