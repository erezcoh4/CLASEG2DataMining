```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
sigma = np.array([0 , 60, 100 , 155 , 175 , 180])
sigma_err = np.array([0 , 10, 20 , 15 , 15 , 15])

sigma2 = np.square(sigma)
sigma2_err = 2*sigma*sigma_err


a2 = np.array([1, 2.2, 3.6 , 4.7 , 5.15 , 5.05])
a2_err = np.array([0 , 0.1 , 0.1 , 0.25 , 0.25 , 0.25])
```

```python
plt.errorbar(a2,sigma,xerr=a2_err,yerr=sigma_err,fmt='o')
p = np.poly1d(np.polyfit(x=a2,y=sigma,deg=1))
plt.plot(a2,p(a2),'--')
```

```python
plt.errorbar(a2,sigma2,xerr=a2_err,yerr=sigma2_err,fmt='o')
p = np.poly1d(np.polyfit(x=a2,y=sigma2,deg=1))
plt.plot(a2,p(a2),'--')
```
