## a general A-dependence plot

```python
def plot_A_dependence_parameter( parlabel = '$a_1$',
                           par_data_values = [0.126,0.081,0.212,0.103],
                           par_data_errors = [0.050 , 0.057 , 0.032 , 0.079],
                           figsize=(12,8),fontsize=25):
    A = [12,27,56,208]
    fig,ax=plt.subplots(figsize=figsize)
#     ax.set_xscale('log')
    plt.errorbar(A,y=par_data_values,yerr=par_data_errors,fmt='o',markersize=10, color='blue' , ecolor='red', capthick=2)
    set_axes(ax,'A',parlabel,fontsize=fontsize)
    ax.xaxis.set_ticks([12,27,56,208])
    ax.grid(linestyle='--',alpha=0.8)
```

```python
plot_A_dependence_parameter(parlabel = '$a_1$',
                           par_data_values = [0.126,0.081,0.212,0.103],
                           par_data_errors = [0.050 , 0.057 , 0.032 , 0.079])
```

```python
plot_A_dependence_parameter(parlabel = '$a_2$ [GeV/c]',
                           par_data_values = [0.160,0.149,0.164,0.181],
                           par_data_errors = [0.007,0.008,0.005,0.010])
```
