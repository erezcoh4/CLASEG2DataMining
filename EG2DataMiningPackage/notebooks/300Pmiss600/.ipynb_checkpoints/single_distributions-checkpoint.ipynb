# $\chi2$-test of data and simulation

```python
columns = ['Pmiss3Mag','pcmX','pcmY','pcmZ'] 
h_data = np.array([0,  0,  1,  0,  3, 11,  9,  6,  6,  5,  1,  1,  1,  0,  1])
h_sim = dict()
h_sim['chi2red_0.6'] = np.array([0,  0,  0,  0,  1, 10, 13,  8, 12,  1,  0,  0,  0,  0,  0])
h_sim['chi2red_1.1'] = np.array([0,  2,  3,  4,  4,  8, 15,  7,  1,  0,  0,  1,  0,  0,  0])
h_sim['chi2red_1.9'] = np.array([0,  0,  2,  8, 14, 11,  9,  1,  0,  0,  0,  0,  0,  0,  0])
h_sim['chi2red_1.6'] = np.array([0,  0,  0,  0,  0, 21, 22,  2,  0,  0,  0,  0,  0,  0,  0])
h_sim['chi2red_2.3'] = np.array([0,  2,  4,  5, 17, 12,  4,  1,  0,  0,  0,  0,  0,  0,  0])
h_sim['chi2red_3.6'] = np.array([0,  0,  0,  0,  0,  0,  0,  0,  2,  3,  5,  5,  7, 10,  9])
```

```python
bins=np.linspace(-0.4,0.8,16)
fig=plt.figure(figsize=(20,20))
for i,chi2 in enumerate([0.6,1.1,1.6,1.9,2.3,3.6]):
    ax = fig.add_subplot(3,2,i+1)
    plt.step(bins[:-1],h_data,label='data',linewidth=3)
    plt.step(bins[:-1],h_sim['chi2red_%.1f'%chi2],label=r'sim.',linewidth=3)
    set_axes(ax,r'$p_{c.m.}^{z}$ [GeV/c]' if i>3 else '','',do_add_legend=True,title=r'$\chi^2_{red}(z)=%.1f$'%chi2)
    
plt.tight_layout()
plt.savefig(figures_path+'chi2red_z_examples.pdf')
```

```python
from scipy.stats import norm
bins=np.linspace(-0.8,0.8,50)

fig=plt.figure(figsize=(16,12))
for i,(filename,chi2,color) in enumerate(zip(['SigmaT0.117_SigmaZ0.202_MeanZ-0.004_chi2_1.1'
                                              ,'SigmaT0.261_SigmaZ0.062_MeanZ-0.036_chi2_8'
                                              ,'SigmaT0.133_SigmaZ0.123_MeanZ0.270_chi2_13'
                                              ,'SigmaT0.206_SigmaZ0.084_MeanZ0.116_chi2_20'                    
                                              ,'SigmaT0.178_SigmaZ0.183_MeanZ0.247_chi2_30'                         
                                              ,'SigmaT0.252_SigmaZ0.167_MeanZ0.268_chi2_72']
                                             ,[1.1,8,13,20,30,72]
                                             ,['forestgreen','blue','orange','red','purple','black'])):
    ax = fig.add_subplot(2,3,i+1)
    title = r'$\chi_{red}^2 \simeq %.0f$'%chi2
    sample = reconstructed[filename]
    plt.hist(sample['pcmX'],histtype='step',color=color,linewidth=3,bins=50,normed=1)
    mean = np.mean(sample['pcmX'])
    sigma = np.std(sample['pcmX'])
    x = np.linspace(np.min(sample['pcmX']),np.max(sample['pcmX']),1000)
    Gauss_x = norm.pdf(x,mean,sigma)
    plt.plot(x,Gauss_x,'--',color=color)
    plt.plot([0,0],[0,0],color=color,linewidth=3,label=label)
    set_axes(ax,'$p_{cm}^{x}$ [GeV/c]' if i>2 else ''
             ,title=title,nticklabels=3
             ,xticks=[float(int(np.min(sample['pcmX'])*10))/10,0,-float(int(np.min(sample['pcmX'])*10))/10])
    ax.yaxis.set_major_formatter(NullFormatter())
plt.tight_layout()
plt.savefig(figures_path+'examples_chi2.pdf')
```
