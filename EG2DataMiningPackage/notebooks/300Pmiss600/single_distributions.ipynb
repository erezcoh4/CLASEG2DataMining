```python
from scipy.stats import norm
bins=np.linspace(-0.8,0.8,50)

fig=plt.figure(figsize=(28,6))
for i,(filename,chi2,color) in enumerate(zip(['SigmaT0.117_SigmaZ0.202_MeanZ-0.004_chi2_1.1'
                                  ,'SigmaT0.261_SigmaZ0.062_MeanZ-0.036_chi2_8'
                                  ,'SigmaT0.206_SigmaZ0.084_MeanZ0.116_chi2_20'                    
                                  ,'SigmaT0.178_SigmaZ0.183_MeanZ0.247_chi2_30'                         
                                  ,'SigmaT0.252_SigmaZ0.167_MeanZ0.268_chi2_72']
                                  ,[1.1,8,20,30,72]
                                  ,['black','blue','red','forestgreen','purple'])):
    ax = fig.add_subplot(1,5,i+1)
    title = r'$\chi_{red}^2 \simeq %.0f$'%chi2
    sample = reconstructed[filename]
    plt.hist(sample['pcmX'],histtype='step',color=color,linewidth=3,bins=50,normed=1)
    mean = np.mean(sample['pcmX'])
    sigma = np.std(sample['pcmX'])
    x = np.linspace(np.min(sample['pcmX']),np.max(sample['pcmX']),1000)
    Gauss_x = norm.pdf(x,mean,sigma)
    plt.plot(x,Gauss_x,'--',color=color)
    plt.plot([0,0],[0,0],color=color,linewidth=3,label=label)
    set_axes(ax,'$p_{cm}^{x}$ [GeV/c]',title=title,nticklabels=3
             ,xticks=[float(int(np.min(sample['pcmX'])*10))/10,0,-float(int(np.min(sample['pcmX'])*10))/10])
    ax.yaxis.set_major_formatter(NullFormatter())
plt.savefig(figures_path+'examples_chi2.pdf')
```
